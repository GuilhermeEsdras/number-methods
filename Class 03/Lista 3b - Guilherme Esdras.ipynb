```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-poster')
```

```python
# "Print Utils"

def print_formatted(formula_name, text, val):
    print(f'----- \n {formula_name} {text}: \n', val)
    return

def separador():
    print('-----')
    return
```

<div class="alert alert-block alert-info">
    Aproximação por séries de Taylor para o arcotangente.
</div>

$$
    \arctan(x) = x - \dfrac{x^3}{3} + \dfrac{x^5}{5} - \dfrac{x^7}{7} + \cdots = \sum^\infty_{n=0} (-1)^n \dfrac{x^{2n+1}}{2n+1}
$$

```python
def arctanT(x, n):
    return ( (-1)**n * ( (x**(2*n+1)) / (2*n+1)) )
```

---

<div class="alert alert-block alert-info">
    a) <b>Fórmula de Machin:</b>
</div>


$$
 \dfrac{\pi}{4} = 4 \ arctan\dfrac{1}{5} - \ arctan\dfrac{1}{239}
$$


```python

machin = 0
machin_lista = []
for i in range(20):
    '''
        Fórmula de Machin
    '''
#     machin += 4 * arctanT((1/5), i) - arctanT((1/239), i)
    machin = 4 * (4 * arctanT((1/5), i) - arctanT((1/239), i))
    
    machin_lista.append(machin)
    print(f'Usando {i} termo: {machin}')

# print(f'\nUsando a Numpy: ', np.arctan(1), '\n')


# ------


machin_arr = np.array(machin_lista)
print_formatted('Machin', 'arr', machin_arr)

machin_diff = np.diff(machin_arr)
print_formatted('Machin', 'diff', machin_diff)

machin_erro = np.abs(machin_diff / machin_arr[0:19])
print_formatted('Machin', 'erro', machin_erro)


# ------
separador()
# ------

machin_x = np.arange(0, 19)

# ------

plt.plot(machin_x, machin_erro, 'r', label = 'Fórmula de Machin')
plt.title("Evolução de Erro - Fórmula de Machin")
plt.xticks(machin_x, machin_x)
plt.xlabel('$Termos$')
plt.ylabel('$Erros$')
plt.grid()
plt.legend()
```

<div class="alert alert-block alert-info">
    b) <b>Fórmula de Hutton:</b>
</div>


$$
 \dfrac{\pi}{4} = arctan\dfrac{1}{2} + arctan\dfrac{1}{3}
$$


```python

hutton = 0
hutton_lista = []

for i in range(20):
    '''
        Fórmula de Hutton
    '''
#     hutton += arctanT((1/2), i) + arctanT((1/3), i)
    hutton = 4 * ( arctanT((1/2), i) + arctanT((1/3), i) )
    
    hutton_lista.append(hutton)
    print(f'Usando {i} termo: {hutton}')

# print(f'\nUsando a Numpy: ', np.arctan(1), '\n')


# ------


hutton_arr = np.array(hutton_lista)
print_formatted('Hutton', 'val', hutton_arr)

hutton_diff = np.diff(hutton_arr)
print_formatted('Hutton', 'diff', hutton_diff)

hutton_erro = np.abs(hutton_diff / hutton_arr[0:19])
print_formatted('Hutton', 'erro', hutton_erro)


# ------
separador()
# ------

hutton_x = np.arange(0, 19)

# ------

plt.plot(hutton_x, hutton_erro, 'g', label = 'Fórmula de Hutton')
plt.title("Evolução de Erro - Fórmula de Hutton")
plt.xticks(hutton_x, hutton_x)
plt.xlabel('Termos')
plt.ylabel('Erros')
plt.grid()
plt.legend()
```

<div class="alert alert-block alert-info">
    c) <b>Fórmula de Clausen:</b>
</div>


$$
 \dfrac{\pi}{4} = 2 \ arctan\dfrac{1}{3} + arctan\dfrac{1}{7}
$$


```python

clausen = 0
clausen_lista = []
for i in range(20):
    '''
        Fórmula de Clausen
    '''
#     clausen += 2 * arctanT((1/3), i) + arctanT((1/7), i)
    clausen = 4 * (2 * arctanT((1/3), i) + arctanT((1/7), i))
    
    clausen_lista.append(clausen)
    print(f'Usando {i} termo: {clausen}')

# print(f'\nUsando a Numpy: ', np.arctan(1), '\n')


# ------


clausen_arr = np.array(clausen_lista)
print_formatted('Clausen', 'arr', clausen_arr)

clausen_diff = np.diff(clausen_arr)
print_formatted('Clausen', 'diff', clausen_diff)

clausen_erro = np.abs(clausen_diff / clausen_arr[0:19])
print_formatted('Clausen', 'erro', clausen_erro)


# ------
separador()
# ------

clausen_x = np.arange(0, 19)

# ------

plt.plot(clausen_x, clausen_erro, 'b', label = 'Fórmula de Clausen')
plt.title("Evolução de Erro - Fórmula de Clausen")
plt.xticks(clausen_x, clausen_x)
plt.xlabel('$Termos$')
plt.ylabel('$Erros$')
plt.grid()
plt.legend()
```

<div class="alert alert-block alert-info">
    d) <b>Fórmula de Dase:</b>
</div>


$$
 \dfrac{\pi}{4} = arctan\dfrac{1}{2} + arctan\dfrac{1}{5} + arctan\dfrac{1}{8}
$$


```python
dase = 0
dase_lista = []
for i in range(20):
    '''
        Fórmula de Dase
    '''
#     dase += arctanT((1/2), i) + arctanT((1/5), i) + arctanT((1/8), i)
    dase = 4 * (arctanT((1/2), i) + arctanT((1/5), i) + arctanT((1/8), i))
    
    dase_lista.append(dase)
    print(f'Usando {i} termo: {dase}')

# print(f'\nUsando a Numpy: ', np.arctan(1), '\n')


# ------


dase_arr = np.array(dase_lista)
print_formatted('Dase', 'arr', dase_arr)

dase_diff = np.diff(dase_arr)
print_formatted('Dase', 'diff', dase_diff)

dase_erro = np.abs(dase_diff / dase_arr[0:19])
print_formatted('Dase', 'erro', dase_erro)


# ------
separador()
# ------

dase_x = np.arange(0, 19)

# ------

plt.plot(dase_x, dase_erro, 'y', label = 'Fórmula de Dase')
plt.title("Evolução de Erro - Fórmula de Dase")
plt.xticks(dase_x, dase_x)
plt.xlabel('$Termos$')
plt.ylabel('$Erros$')
plt.grid()
plt.legend()
```
